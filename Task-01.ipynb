```python
import numpy as np
import pandas as pd
from pandas import DataFrame,Series
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv(r"C:\Users\chinm\OneDrive\Desktop\Data\adult1.csv")
df = pd.DataFrame(data)
df.head()
```

```python
fig = plt.figure(figsize = (10,5))
ax = fig.add_axes([0,0,0.5,0.8])
x = list(df['sex'])
y = list(df['capital_gain'])
ax.bar(x,y,color='b',width = 0.50)
plt.show()
```

```python
workclass = list(df['workclass'])
hours_per_week = list(df['hours_per_week'])

flg = plt.figure(figsize = (12,6))

plt.bar(workclass, hours_per_week, color = 'yellow', width = 0.50)

plt.xlabel("Mode of workclass")
plt.ylabel("hours_per_week")
plt.title("hours of work per week based om the workclass.")
plt.show()
```

```python
fig , ax = plt.subplots(figsize=(10,7))

sns.histplot(data=df, x='age', bins=30,kde=True, color='blue')

plt.title("Age Histogram")
plt.xlabel('Age')
plt.ylabel('count')
plt.show()
```
